```python
import numpy as np
# Specify the file path
LRO_file_path = 'Al0.20Co0.20Cr0.20Fe0.20Ni0.20/LRO.txt'
Al_file_path = 'Al0.20Co0.20Cr0.20Fe0.20Ni0.20/Al.txt'
Co_file_path = 'Al0.20Co0.20Cr0.20Fe0.20Ni0.20/Co.txt'
Cr_file_path = 'Al0.20Co0.20Cr0.20Fe0.20Ni0.20/Cr.txt'
Fe_file_path = 'Al0.20Co0.20Cr0.20Fe0.20Ni0.20/Fe.txt'
Ni_file_path = 'Al0.20Co0.20Cr0.20Fe0.20Ni0.20/Ni.txt'



# Use numpy.loadtxt to read the file into a NumPy array
LRO_array = np.loadtxt(LRO_file_path)
Al_array = np.loadtxt(Al_file_path)
Co_array = np.loadtxt(Co_file_path)
Cr_array = np.loadtxt(Cr_file_path)
Fe_array = np.loadtxt(Fe_file_path)
Ni_array = np.loadtxt(Ni_file_path)


#print("Original data array:")
#print(len(LRO_array))
import matplotlib.pyplot as plt

# Generate some sample data
T = list(range(1800, 0, -1))  # Create an array of 100 values from 0 to 10
x = np.array(T)

y1 = Al_array
y2 = Co_array
y3 = Cr_array
y4 = Fe_array
y5 = Ni_array


# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y1, label='Al')
ax.plot(x, y2, label='Co')
ax.plot(x, y3, label='Cr')
ax.plot(x, y4, label='Fe')
ax.plot(x, y5, label='Ni')



#plt.ylim(0, 1.0)
#plt.xlim(0, 1800)

# Add labels and a legend
ax.set_xlabel('Temperature (K)',fontdict={'fontsize': 16, 'fontweight': 'bold', 'fontfamily': 'serif'})
ax.set_ylabel('LRO',fontdict={'fontsize': 16, 'fontweight': 'bold', 'fontfamily': 'serif'})
ax.set_title('Al20Co20Cr20Fe20Ni20',fontdict={'fontsize': 22, 'fontweight': 'bold', 'fontfamily': 'serif'})
ax.legend()

plt.xticks(fontsize=12, fontweight='bold', fontfamily='serif')
plt.yticks(fontsize=12, fontweight='bold', fontfamily='serif')
plt.legend(prop={'size': 12, 'weight': 'bold', 'family': 'serif'})
# Show the plot
plt.show()
```
